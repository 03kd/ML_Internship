#####  Q-4

```python
df =data1.copy()
df.drop(columns='index',inplace =True)
```

```python
df['Month'] = pd.DatetimeIndex(df['Date']).month
df['Year'] = pd.DatetimeIndex(df['Date']).year
```

```python
df
```

```python
df1=df.groupby(['Year','Month'],axis=0).mean()
df1
```

```python
df1['VWAP']=(df1['Average Price']*df1['No. of Trades']).sum()/df1['Total Traded Quantity']
```
