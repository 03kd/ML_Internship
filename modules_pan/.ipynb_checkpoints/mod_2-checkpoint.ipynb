## 2.10 Trade Calls - Using Bollinger Bands
## Plot the bollinger bands for this stock - the duration of 14 days and 2 standard deviations away from the average
## The bollinger bands comprise the following data points-

    The 14 day rolling mean of the closing price (we call it the average) 
    Upper band which is the rolling mean + 2 standard deviations away from the average. 
    Lower band which is the rolling mean - 2 standard deviations away from the average. 
    Average Daily stock price.

Bollinger bands are extremely reliable , with a 95% accuracy at 2 standard deviations , and especially useful in sideways moving market.
Observe the bands yourself , and analyse the accuracy of all the trade signals provided by the bollinger bands.
Save to a new csv file.

```python
dataset.head()
```

```python
roll_14 = dataset['Close Price'].rolling(14).mean()[13:]
```

```python
std_14 = dataset['Close Price'].rolling(14).std()[13:]
```

```python
upper_band = roll_14 + 2*std_14
```

```python
lower_band = roll_14 - 2*std_14
```

```python
Avg_stock_price = dataset['Average Price'][13:]
```

```python
bollinger_analysis = pd.DataFrame({'Date' : dataset.Date[13:],
                                    'Avg_stock_price' : Avg_stock_price,
                                    'Rolling_14' : roll_14,
                                    'std_14' : std_14,
                                    'upper_band' : upper_band,
                                    'lower_band' : lower_band})
```

```python
bollinger_analysis.head()
```

```python
sns.lineplot(data = bollinger_analysis, y = 'Rolling_14', x = 'Date', label = '14 days rolling mean', color = 'Orange')
sns.lineplot(data = bollinger_analysis, y = 'upper_band', x = 'Date', label = 'upper band', color = 'Navy')
sns.lineplot(data = bollinger_analysis, y = 'lower_band', x = 'Date', label = 'lower band', color = 'blue')
sns.lineplot(data = bollinger_analysis, y = 'Avg_stock_price', x = 'Date', label = 'Avg Stock Price', color = 'limegreen')
plt.gca().set(xlabel = 'Dates', ylabel = 'Stock Price', title = 'Bollinger Bands')
```

The bands tend to be quite accurate in making a prediction of volatility. It can be clearly seen that whenever the upper and lower bands squeeze or narrow down, there is relatively a situation of low volatility but when these bands move apart the change in stock price is very volatile as evident from the plot.  
Also, a squeeze is followed by broadening out of the bands and vice versa which can give a hint about what trend the stock price is about to follow in the future. 

```python
dataset['rolling_14'] = dataset['Close Price'].rolling(14).mean()
dataset['std_14'] = dataset['Close Price'].rolling(14).std()
dataset['upper_band'] = dataset.rolling_14 + 2*dataset.std_14
dataset['lower_band'] = dataset.rolling_14 - 2*dataset.std_14
```
