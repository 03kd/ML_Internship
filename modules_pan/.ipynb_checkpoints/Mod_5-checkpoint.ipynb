### 5.3 Prepare a scatter plot for differing weights of the individual stocks in the portfolio , the axes being the returns and volatility. Colour the data points based on the Sharpe Ratio ( Returns/Volatility) of that particular portfolio.

Defining a function that would randomly assign weights to the five stocks and calculate the net returns and net volatility in the form of a list of tuples.

```python
def returns_volat(com_returns, com_vol, n):
    rv = []
    for i in range(n):
        a = np.random.random(5)
        a /= a.sum()
        rv.append((np.dot(com_returns,a), np.dot(com_vol,a)))
    return rv
```

```python
d = returns_volat(com_returns, com_vol, 125000)
```

```python
returns = [k[0] for k in d ]
```

```python
volatility = [k[1] for k in d ]
```

```python
sharpe = [k[0]/k[1] for k in d]
```

```python
plt.rcParams['figure.figsize'] = 16,7
sea.set_style('darkgrid')
cmap = sea.cubehelix_palette(as_cmap=True)
plt.scatter(volatility, returns, c=sharpe, cmap = 'GnBu' )
plt.colorbar(label = 'sharpe ratio')
plt.gca().set(xlabel = '% Volatility', ylabel = '% Returns' )
```

### 5.4 Mark the 2 portfolios where -
Portfolio 1 - The Sharpe ratio is the highest  
Portfolio 2 - The volatility is the lowest. 

```python
plt.scatter(volatility, returns, c = sharpe, cmap = 'GnBu')
plt.colorbar(label = 'sharpe ratio')
plt.gca().set(xlabel = '% Volatility', ylabel = '% Returns' )

# Highest Sharpe-ratio
plt.plot(volatility[np.argsort(sharpe)[-1]], returns[np.argsort(sharpe)[-1]], marker = "v", markerfacecolor = 'White')

# Lowest Volatility
plt.plot(volatility[np.argsort(volatility)[0]], returns[np.argsort(volatility)[0]], marker = "v", markerfacecolor = 'Crimson')
```
