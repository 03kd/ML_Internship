### 4.1 Import the csv file of the stock which contained the Bollinger columns as well.
### Create a new column 'Call' , whose entries are -
 - 'Buy' if the stock price is below the lower Bollinger band 
 - 'Hold Buy/ Liquidate Short' if the stock price is between the lower and middle Bollinger band 
 - 'Hold Short/ Liquidate Buy' if the stock price is between the middle and upper Bollinger band 
 - 'Short' if the stock price is above the upper Bollinger band


### Now train a classification model with the 3 bollinger columns and the stock price as inputs and 'Calls' as output.
### Check the accuracy on a test set. (There are many classifier models to choose from, try each one out and compare the accuracy for each)

#### Function for new column 'call'

```python

def call(data):

    middle_band = data.middle_band[13:].values
    upper_band = data.upper_band[13:].values
    lower_band = data.lower_band[13:].values
    stock_price = data['Close Price'][13:].values

    calls = ['NA']*13
    for i in range(len(stock_price)):
        if stock_price[i] < lower_band[i]:
            calls.append('Buy')
        elif stock_price[i] >= lower_band[i] and stock_price[i] < middle_band[i]:
            calls.append('Hold Buy/ Liquidate Short')
        elif stock_price[i] < upper_band[i] and stock_price[i] >= middle_band[i]:
            calls.append('Hold Short/ Liquidate Buy')
        else:
            calls.append('Short')
    return calls
```

```python
data['calls'] = call(data)
```

### 4.2 Now, we'll again utilize classification to make a trade call, and measure the efficiency of our trading algorithm over the past two years. For this assignment , we will use RandomForest classifier.
### - Import the stock data file of your choice
### - Define 4 new columns , whose values are:     
 - % change between Open and Close price for the day   
 - % change between Low and High price for the day   
 - 5 day rolling mean of the day to day % change in Close Price   
 - 5 day rolling std of the day to day % change in Close Price  
### - Create a new column 'Action' whose values are:   
 - 1 if next day's price(Close) is greater than present day's.   
 - (-1) if next day's price(Close) is less than present day's.   
 - i.e. Action [ i ] = 1 if Close[ i+1 ] > Close[ i ]   
 - i.e. Action [ i ] = (-1) if Close[ i+1 ] < Close[ i ]  
### - Construct a classification model with the 4 new inputs and 'Action' as target
### - Check the accuracy of this model , also , plot the net cumulative returns (in %) if we were to follow this algorithmic model


```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
```

```python
rfc = RandomForestClassifier(random_state = 1,
                                  n_estimators = 1000,
                                  max_depth = 9, 
                                  min_samples_split = 2,  min_samples_leaf = 1)
```

```python
data
```

```python
perc_change_open_close = (data['Close Price'].values/data['Open Price'].values - 1)*100
```

```python
perc_change_low_high = (data['High Price'].values/data['Low Price'].values - 1)*100
```

```python
data['perc_change_open_close'] = perc_change_open_close
data['perc_change_low_high'] = perc_change_low_high
data['rolling_5'] = data['Close Price'].rolling(5).mean()
data['std_5'] = data['Close Price'].rolling(5).std()
```

#### Defining the action column:

```python
def action(data):
    action = []
    for i in range(len(data)-1):
        if data['Close Price'][i] < data['Close Price'][i+1]:
            action.append(1)
        else:
            action.append(-1)
    return action
```

```python
data['Action'] = action(data) +['NA']
```

#### As the last row won't have the Action cloumn data and the first 4 rows due to rolling avg. 

```python
X = data[['perc_change_open_close', 'perc_change_low_high', 'rolling_5', 'std_5']].values[4:-1]
```

#### Cummulative % Returns

```python
pred_actions = tuned_rfc.predict(X)
```

```python
pred_actions = [d  if d == 1 else -1 for d in pred_actions]
```

```python
def c_returns(day_wise_returns, pred_actions):
    returns = []
    for i in range(5,len(pred_actions)):
        if pred_actions[i] == 1:
            returns.append(day_wise_returns[i])
        else:
            returns.append(-day_wise_returns[i])
    return returns
            
    
```

Calculating cummulative returns

```python
day_wise_returns = (data['Close Price'].values[5:]/data['Close Price'].values[5]-1)
```

```python
day_wise_returns[:10]
```

```python
day_wise_returns = list(day_wise_returns)
```

```python
data['day_wise_returns'] = ([0]*4 + day_wise_returns + [0])
```

```python
data['pred_actions'] = ['NA']*4 + pred_actions + ['NA']
```

```python
algo_cummulative_returns = data.pred_actions.values[4:-1]*data.day_wise_returns.values[4:-1]
```

```python
algo_cummulative_returns =[0]*4 + list(algo_cummulative_returns) + [0]
```

```python
len(algo_cummulative_returns)
```

```python
data['algo_cummulative_returns'] = algo_cummulative_returns
```

```python
data['algo_cummulative_returns'] = data['algo_cummulative_returns'].cumsum()
```
