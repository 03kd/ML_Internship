```python
df1 = df.groupby(['Trend'], as_index=False)
```

### Q-10

```python
df.head()
```

```python
roll_14 = df['Close Price'].rolling(14).mean()[13:]
std_14  = df['Close Price'].rolling(14).std()[13:]
```

```python
upper_band = roll_14 + 2*std_14
lower_band = roll_14 - 2*std_14
```

```python
avg_stock_price = df['Average Price'][13:]
```

```python
bollinger_df = pd.DataFrame({'Date':avg_stock_price.index.values,
                            'Rolling_14':roll_14, 
                            'Std_14':std_14,
                             'Average Price':avg_stock_price,
                            'Upper_band':upper_band,
                            'Lower_band':lower_band})
```

```python
bollinger_df.head()
```

```python
sns.lineplot(data =bollinger_df, x ='Date', y = 'Average Price', label ='Average Price', color='DarkGreen')
sns.lineplot(data =bollinger_df, x ='Date', y = 'Rolling_14', label ='Rolling 14', color='Maroon')
sns.lineplot(data = bollinger_df, x ='Date', y = 'Upper_band', label ='Upper Band', color='Black')
sns.lineplot(data = bollinger_df, x ='Date', y = 'Lower_band', label ='Lower Band', color='DodgerBlue')
plt.gca().set(xlabel ='Date', ylabel ='Stock Price', title ='Bollinger Bands')
```

The bands tend to be quite accurate in making a prediction of volatility. It can be clearly seen that whenever the upper and lower bands squeeze or narrow down, there is relatively a situation of low volatility but when these bands move apart the change in stock price is very volatile as evident from the plot.
Also, a squeeze is followed by broadening out of the bands and vice versa which can give a hint about what trend the stock price is about to follow in the future. 

```python
df['Rolling_14'] = df['Close Price'].rolling(14).mean()
df['Std_14'] = df['Close Price'].rolling(14).std()
df['Upper_band'] = roll_14 + 2*std_14
df['Lower_band'] = roll_14 - 2*std_14
```
